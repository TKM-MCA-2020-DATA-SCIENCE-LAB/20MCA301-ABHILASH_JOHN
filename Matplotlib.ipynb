```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.randn(10)
```

```python
y= np.random.rand(10)
y
```

```python
plt.plot(x,y,'g')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("X-Y Plane")
```

```python
plt.subplot(2,2,1)
plt.plot(x,y*10,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'r*-')
plt.subplot(2,2,3)
plt.plot(x,y,'y+-')
plt.subplot(2,2,4)
plt.plot(x,y**2,'b+-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0.5,0.9,0.9])
axes2 = fig.add_axes([0.07,0.97,0.3,0.3])
axes2.plot(x,y,"r+-")
axes.plot(np.arange(109,119),np.arange(0,10))

```
